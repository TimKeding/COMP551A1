```python
from pca import PCAPlotting
```

1. import merged data
2. remove regions that contain >= 85% of data being NaN
3. fill all NaNs with 0s

```python
from data import Data
data = Data()
data.filter_low_data_regions(threshold=0.15)
data.fill_na(0)
dat = data.d
```

Generate a PCA-reduced dataset and generate an elbow graph for the first 10
dimensions demonstrating the percent of variance each dimension accounts for.

```python
pca = PCAPlotting(dat)
pca.plot_optimal_pc(20)
```

Print percentage of variance covered for each dimension added

```python
variance = 0
for i in range(len(pca.pca.explained_variance_ratio_)):
    variance += pca.pca.explained_variance_ratio_[i] * 100
    print(f"{i+1} Dimension(s) account for {variance}% of the dataset variance." )
```

For visibility, generate a two dimensional and third dimensional graph to
display the PCA reduced data

```python
pca.reduce_dimensionality(2)
pca.plot_data()
pca.reduce_dimensionality(3)
pca.plot_data()
```
