```python
from pca import PCAPlotting
from kmeans import Kmeans
```

1. import merged data
2. remove regions that contain >= 85% of data being NaN
3. fill all NaNs with 0s

```python
from data import Data
data = Data()
data.filter_low_data_regions(threshold=0.15)
data.fill_na(0)
data.normalize_regions()
dat = data.d
```

Generate two PCA-reduced datasets, reducing to the second and third dimension for

```python
two_dim_pca = PCAPlotting(dat)
two_dim_pca.reduce_dimensionality(2)
three_dim_pca = PCAPlotting(dat)
three_dim_pca.reduce_dimensionality(3)
```

Utilize kmeans to calculate clusters for the graphs. Based on optimal k values
determined by elbow graphs, there will be 10 clusters for both graphs.
Generate a two dimensional and third dimensional graph to
display the PCA reduced data

```python
two_dim_kmeans = Kmeans(two_dim_pca.reduced_data)
three_dim_kmeans = Kmeans(three_dim_pca.reduced_data)
two_dim_kmeans.find_clusters(10)
three_dim_kmeans.find_clusters(10)
two_dim_kmeans.plot_kmeans()
three_dim_kmeans.plot_kmeans()
```
